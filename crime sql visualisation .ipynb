Q4= Primary_crime type and their total no of crimes  with their total percentage .

```python
q4 = """
SELECT
    `Primary_Type`,
    COUNT(*) AS total_crimes,
    ROUND((COUNT(*) / (SELECT COUNT(*) FROM crime_details)) * 100, 2) AS percentage
FROM crime_details
GROUP BY `Primary_Type`
ORDER BY percentage DESC;
"""

cursor = mydb.cursor()
cursor.execute(q4)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```

q5=What's the average response time between Case_Time and Update_Time for different crime types?

```python
q5=''' 
SELECT 
    Primary_Type,
    AVG(TIMESTAMPDIFF(MINUTE, 
        CONCAT(Case_Date, ' ', Case_Time),
        CONCAT(Update_Date, ' ', Update_Time))) as avg_response_minutes
FROM crime_details
GROUP BY Primary_Type
HAVING avg_response_minutes > 0
ORDER BY avg_response_minutes DESC;
'''

cursor = mydb.cursor()
cursor.execute(q5)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```

q6= Which beats (police areas) have shown the most improvement (reduction in crime) over the years?

```python
q6=''' 
WITH yearly_beat_crimes AS (
    SELECT 
        Beat,
        Year,
        COUNT(*) as crime_count,
        LAG(COUNT(*)) OVER (PARTITION BY Beat ORDER BY Year) as prev_year_count
    FROM crime_details
    GROUP BY Beat, Year
)
SELECT 
    Beat,
    Year,
    crime_count,
    prev_year_count,
    ROUND(((crime_count - prev_year_count) * 100.0) / prev_year_count, 2) as change_percentage
FROM yearly_beat_crimes
WHERE prev_year_count IS NOT NULL
ORDER BY change_percentage;
'''

cursor = mydb.cursor()
cursor.execute(q6)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```

q7=What percentage of crimes occur during weekends vs weekdays?

```python
q7=''' 
SELECT 
    CASE 
        WHEN DAYOFWEEK(Case_Date) IN (1,7) THEN 'Weekend'
        ELSE 'Weekday'
    END as day_type,
    COUNT(*) as crime_count,
    ROUND((COUNT(*) * 100.0) / (SELECT COUNT(*) FROM crime_details), 2) as percentage
FROM crime_details
GROUP BY day_type;
'''

cursor = mydb.cursor()
cursor.execute(q7)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```

q8= Total number of crimes every season .

```python
#  q8= Total number of crimes every season .
q8 = """
SELECT
    CASE
        WHEN MONTH(Case_Date) IN (12, 1, 2) THEN 'Winter'
        WHEN MONTH(Case_Date) IN (3, 4, 5) THEN 'Spring'
        WHEN MONTH(Case_Date) IN (6, 7, 8) THEN 'Summer'
        ELSE 'Autumn'
    END AS season,
    COUNT(*) AS total_crimes
FROM crime_details
GROUP BY season
ORDER BY FIELD(season, 'Winter', 'Spring', 'Summer', 'Autumn');
"""

cursor = mydb.cursor()
cursor.execute(q8)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```

q9=Total no of crimes in every season and wheather it is occured in domestic or non_domestic?

```python
q9 = """
SELECT
    CASE
        WHEN MONTH(Case_Date) IN (12, 1, 2) THEN 'Winter'
        WHEN MONTH(Case_Date) IN (3, 4, 5) THEN 'Spring'
        WHEN MONTH(Case_Date) IN (6, 7, 8) THEN 'Summer'
        ELSE 'Autumn'
    END AS season,
    COUNT(*) AS total_crimes,
    SUM(CASE WHEN Domestic = 1 THEN 1 ELSE 0 END) AS domestic_crimes,
    SUM(CASE WHEN Domestic = 0 THEN 1 ELSE 0 END) AS non_domestic_crimes
FROM crime_details
GROUP BY season
ORDER BY FIELD(season, 'Winter', 'Spring', 'Summer', 'Autumn');
"""

cursor = mydb.cursor()
cursor.execute(q9)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```
