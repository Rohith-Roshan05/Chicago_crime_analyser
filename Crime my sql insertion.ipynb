```python
import pandas as pd
import mysql.connector
import time

# Connect to the database
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",  # Replace with your MySQL password
    database="chicago_crime_details",
    port=3307
)
my_cursor = mydb.cursor()

# Load and clean CSV data
csv_file = r"Z:\Chicago project\crime_cleaned.csv"
crime_cleaned = pd.read_csv(csv_file)

# Remove duplicates and empty rows
crime_cleaned = crime_cleaned.drop_duplicates()
crime_cleaned = crime_cleaned.dropna()

# Check the cleaned row count
print(f"Rows after cleaning: {len(crime_cleaned)}")

# Prepare the insert query
insert_query = '''
INSERT INTO crime_details(
    ID, Case_Number, Case_Date, Case_Time, Year, Block, IUCR, Primary_Type,
    Description, Location_Description, Arrest, Domestic, Beat, District,
    FBI_Code, Update_Date, Update_Time, Latitude, Longitude, Location
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
'''

# Clear the table before inserting new data
my_cursor.execute("TRUNCATE TABLE crime_details")

# Insert data in batches
batch_size = 5000
data = crime_cleaned.values.tolist()

for i in range(0, len(data), batch_size):
    batch = data[i:i + batch_size]
    try:
        start_time = time.time()
        my_cursor.executemany(insert_query, batch)
        mydb.commit()
        end_time = time.time()
        print(f"Batch {i // batch_size + 1} inserted in {end_time - start_time:.2f} seconds")
    except Exception as e:
        print(f"Error in batch {i // batch_size + 1}: {e}")

print("Data insertion complete.")
my_cursor.close()
mydb.close()
```
